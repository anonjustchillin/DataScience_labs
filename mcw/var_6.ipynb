```python
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
import numpy as np
```

Білет №6
Реалізувати скрипт із згладжуванням вибірки вимірів за методом найменших квадратів.

```python
def generate_data(n=100, a=0.4, b=7, c=10, scale=20):
    x = np.linspace(0, 20, n)
    y_trend = a*x**2 + b*x + c
    S = np.random.exponential(scale=scale, size=n)
    y = y_trend + (S - scale)
    return x, y
```

```python
x,y = generate_data()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(x, y, color='blue')
plt.title('Згенерована вибірка')
plt.grid(True, alpha=0.4)
plt.show()
```

```python
plt.hist(y, edgecolor = "black")
plt.title('Розподіл значень y')
plt.show()
```

```python
def plot_smooth(x, y1, y2, name=''):
    plt.figure(figsize=(12, 5))
    plt.plot(x,y1, color='blue', label='original')
    plt.plot(x,y2, color='red', label='smoothed')
    plt.title(f'Згладжена вибірка ({name})')
    plt.grid(True, alpha=0.4)
    plt.legend()
    plt.show()
```

```python
win_len=len(x)
polyorder=2
y_smooth = savgol_filter(y, window_length=win_len, polyorder=polyorder)
plot_smooth(x, y, y_smooth, name=f'Savitzky-Golay: win_len={win_len}, polyorder={polyorder}')
```

```python
win_len=10
polyorder=2
y_smooth = savgol_filter(y, window_length=win_len, polyorder=polyorder)
plot_smooth(x, y, y_smooth, name=f'Savitzky-Golay: win_len={win_len}, polyorder={polyorder}')
```

```python
win_len=len(x)
polyorder=1
y_smooth = savgol_filter(y, window_length=win_len, polyorder=polyorder)
plot_smooth(x, y, y_smooth, name=f'Savitzky-Golay: win_len={win_len}, polyorder={polyorder}')
```
